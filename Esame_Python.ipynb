```python
# Carico il file e inizio ad analizzare le dimensioni del dataset e relativi metadati
dati_covid = pd.read_csv('owid-covid-data.csv')
```

```python
# Qui ho stampato le colonne per avere un'idea dei nomi da inserire nelle varie formule
for column in dati_covid.columns:
    print(column)
```

```python
# Casi per continente
casi_per_continente = dati_covid.groupby('continent')['total_cases'].sum()
print(casi_per_continente)
```

```python
# Totale casi mondiali
casi_mondiali = casi_per_continente.sum()
print(casi_mondiali)
```

```python
# Percentuale per continente
percentuale_per_continente = (casi_per_continente / casi_mondiali) * 100
print(percentuale_per_continente)
```

```python
# Evoluzione dei casi totali in italia, nel 2022
italia = dati_covid[(dati_covid['location'] == 'Italy') & (dati_covid['date'].str.startswith('2022'))]
italia = italia[['date', 'total_cases']]
print(italia)
```

```python
# Qui ho risconttrato un problema con i nomi delle colonne e ho risolto semplicemente rinominandole
if 'total_cases' not in italia.columns:
    italia = italia.rename(columns={'new_cases': 'total_cases'})
print(italia)
```

```python
df_italia_giornaliero = italia.groupby('date')['total_cases'].sum().reset_index()
```

```python
# Ho raggruppato le date e le ho messe sull'asse delle x
# I casi totali sulle y
# Ho migliorato l'aspetto visivo del grafico aggiungendo titoli, nomi e grigia

plt.figure(figsize=(10, 6))
df_italia_giornaliero.plot(x='date', y='total_cases', kind='line', color='blue', ax=plt.gca())

plt.title('Andamento dei casi COVID-19 in Italia (2022)', fontsize=16)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Numero di casi', fontsize=12)

plt.xticks(rotation=45)

plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# Somma dei pazienti ospedalizzati per Italia, Germania, Francia e Spagna nel 2023

df_2023 = dati_covid[(dati_covid['location'].isin(['Italy', 'Germany', 'France', 'Spain'])) & (dati_covid['date'].str.startswith('2023'))]
print(df_2023)
```

```python
ospitalizzati = df_2023.groupby('location')['hosp_patients'].sum()
print (ospitalizzati)
```

```python
# Il dataset presenta la colonna dei pazienti ospitalizzati vuota per la Germania, che rappresenta un dato non vero,
# infatti, controllando le altre colonne, notiamo che esistono pazienti ospitalizzati settimanalmente, anche per quanto riguarda la terapia intensiva
# "correggere" questo dato non è possibile a meno che non avessimo proprio la possibilità di richiedere i dati
# quindi, penso che potremmo basarci su altre colonne per arrivare ad avere quanto meno un'idea di quanti ospedalizzati ci siano stati in Germania
```

```python
ospitalizzati_dati_di_confronto = df_2023.groupby('location')['weekly_hosp_admissions'].sum()
print(ospitalizzati_dati_di_confronto)
```

```python
df_boxplot = dati_covid[
    (dati_covid['location'].isin(['Italy', 'Germany', 'France'])) &
    (dati_covid['date'] >= '2022-05-01') &
    (dati_covid['date'] <= '2023-04-30')
]
print(df_boxplot)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data = df_boxplot, x='location', y='icu_patients',)


plt.title('Pazienti in terapia intensiva (Maggio 2022 - Aprile 2023)', fontsize=16)
plt.xlabel('Paese', fontsize=12)
plt.ylabel('Pazienti in terapia intensiva', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
